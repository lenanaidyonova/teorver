### Введем матрицу перехода

```python
import random

m = [[3, 1, 0, 0, 2, 0, 0, 4],
     [2, 2, 4, 0, 2, 0, 0, 0],
     [2, 2, 5, 0, 0, 0, 0, 1],
     [0, 0, 0, 2, 0, 4, 4, 0],
     [2, 3, 0, 0, 3, 0, 0, 2],
     [0, 0, 0, 2, 0, 3, 5, 0],
     [0, 2, 0, 0, 0, 4, 4, 0],
     [4, 0, 0, 0, 4, 0, 0, 2]]
```

### Задача 6

```python
def six(m):
    vertexes = [i for i in range(8)]
    MC_lengths = [10, 50, 100, 1000]
 
    answer = []
 
    for MC_len in MC_lengths:
        for vertex in range(8):
            cur_answer = [str(vertex+1)]
            cur_vertex = vertex
            for i in range(MC_len):
                cur_vertex = random.choices(vertexes, weights=m[cur_vertex])[0]
                cur_answer.append(str(cur_vertex+1))
            answer.append(cur_answer)
 
    for i in range(4):
        print("\033[1m" + f"Трактории для длинны {MC_lengths[i]}" + "\033[0m", end="\n\n")
        for j in range(i * (len(answer) // 4), (i+1) * (len(answer) // 4)):
            print("\033[1m" + f"Начало в cостоянии {answer[j][0]}: " + "\033[0m", "->".join(answer[j]), end="\n\n")
```

```python
six(m)
```

### Задача 7

```python
def seven(m):
    vertexes = [i for i in range(8)]
    predict = [78/290, 45/290, 36/290, 0, 70/290, 0, 0, 61/290]
    answer = []
    iters = [10, 100, 1000]
 
    for iter in iters:
        cur_answer = [0]*8
        for vertex in range(8):
            for i in range(iter):
                cur_vertex = vertex
                for j in range(iter):
                    cur_vertex = random.choices(vertexes, weights=m[cur_vertex])[0]
                cur_answer[cur_vertex] += 1
        answer.append(cur_answer)
 
    for i in range(len(iters)):
        print("\033[1m" + f"{iters[i]} итераций" + "\033[0m")
        for j in range(8):
            print(f"Состояние {j+1} - {answer[i][j] * 1.0 / (8*iters[i])}, ", f"теоритическое значение - {predict[j]*1.0}")
```

```python
seven(m)
```
